```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a random 3D dataset with 3 clusters
X, y = make_blobs(n_samples=300, centers=3, n_features=3, random_state=42)

# Visualize the dataset
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

# Fit GMM to the dataset
gmm = GaussianMixture(n_components=3)
gmm.fit(X)

# Predict the cluster for each point
y_gmm = gmm.predict(X)

# Visualize the GMM clustering
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y_gmm, cmap='viridis')
plt.show()
```

```python
# Compare original labels and GMM-predicted labels
print("Original Labels from make_blobs:\n", y)
print("GMM-Predicted Labels:\n", y_gmm)
```
